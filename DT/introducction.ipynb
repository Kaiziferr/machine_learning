```python
counter = 0
main_dict = {}

for col in cols:
  main_dict[col] = []
```

```python
import urllib3
```

```python
url = 'https://raw.githubusercontent.com/Kaiziferr/datasets/main/diabetes.csv'
```

```python
http = urllib3.PoolManager()
r = http.request('GET', url)
```

```python
print('Status request code {}'.format(r.status))
```

```python
response = r.data
```

```python
# string binarize to string utf-8
str_data = response.decode('utf-8')
```

```python
str_data
```

```python
# Data split to \n
lines = str_data.split('\n')
```

```python
# Columns
columns_names = lines[0].split(',')
n_cols = len(columns_names)
n_cols, columns_names
```

```python
counter = 0
main_dict = {}
for i in range(n_cols):
  main_dict[columns_names[i]] = []

main_dict
```

```python
for line in lines:
  if counter > 0:
    values = line.strip().split(',')
    for i, val in enumerate(values):
      main_dict[columns_names[i]].append(val)
  counter = counter + 1
```

```python
data_medica = pd.DataFrame(main_dict)
```

```python
print('The dataset has %d rows and %d columns'%(counter, n_cols))
```

```python
import os
```

```python
mainpath = './data2'
filename = 'registros_medicos.'
fullpath = os.path.join(mainpath, filename)
```

```python
data_medica.to_csv(f'{fullpath}.csv')
data_medica.to_json(f'{fullpath}.json')
data_medica.to_excel(f'{fullpath}.xlsx')
```
