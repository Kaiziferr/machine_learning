# **Muestreo sistematico**
---
No es posible determinar una función para todos los casos, ya que el muestreo sistematico dependera de la regla que se usara como condición.

```python
def systematic_sampling(data, step):
  indexes = np.arange(0, len(data), step=step)
  systematic_sample = data.iloc[indexes]
  return systematic_sample

systematic_sample = systematic_sampling(data, 3)
systematic_sample
```
