```python
from sklearn.datasets import (make_blobs,
                              make_moons,
                              make_circles,
                              make_regression)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
X, y = make_blobs(n_samples=100, centers=3, n_features=2)
data = pd.DataFrame(X, columns=['A', 'B'])
data['Y'] = y
fig, ax = plt.subplots()
colors = {0:'red', 1:'blue', 2:'red'}
sns.scatterplot(data=data, x="A", y="B", palette="deep", hue='Y')
plt.show()
```

```python
X, y = make_moons(n_samples=100, noise=0.08)
data = pd.DataFrame(X, columns=['A', 'B'])
data['Y'] = y
fig, ax = plt.subplots()
colors = {0:'red', 1:'blue', 2:'red'}
sns.scatterplot(data=data, x="A", y="B", palette="deep", hue='Y')
plt.show()
```

```python
X, y = make_circles(n_samples=1000, noise=0.05)
data = pd.DataFrame(X, columns=['A', 'B'])
data['Y'] = y
fig, ax = plt.subplots()
colors = {0:'red', 1:'blue', 2:'red'}
sns.scatterplot(data=data, x="A", y="B", palette="deep", hue='Y')
plt.show()
```

```python
X, y = make_regression(n_samples=100, n_features=1, noise=2.8)
data = pd.DataFrame(X, columns=['A'])
data['Y'] = y
fig, ax = plt.subplots()
sns.scatterplot(data=data, x="A", y="Y")
plt.show()
```
