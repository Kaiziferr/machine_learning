```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
```

- Suponga que las calificaciones del examen parcial de **50** alumnos en
un curso de estadística tuvieron un promedio de **70** y una desviación
estándar de **5**.

```python
mu, sigma = 70, 5

data = np.random.normal(mu, sigma, 50)
```

```python
sns.histplot(data, bins=10, kde=True, stat='probability')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.title("Distribution of Student's Score")
plt.show()
```

```python
#z-score
data2 = (data-mu)/sigma
```

```python
# empirical rule
first_group = len(data2[(data2>-1) & (data2<1)])/len(data2)*100
second_group = len(data2[(data2>-2) & (data2<2)])/len(data2)*100
three_group = len(data2[(data2>-3) & (data2<3)])/len(data2)*100
```

```python
print('Percentage of data with in one standard deviation:', round(first_group,2))
print('Percentage of data with in two standard deviation:', round(second_group,2))
print('Percentage of data with in three standard deviation:', round(three_group,2))
```

```python
sns.histplot(data2, bins=10, kde=True, stat='probability')
plt.xlabel('Student')
plt.ylabel('Probability')
#plt.xticks(range(-3,4,1))
plt.title("Distribution of Student's Score")
plt.show()
```

- ¿Qué porcentaje de estudiantes obtuvieron menos de 70?

```python
z = (70-mu)/sigma
p = norm.cdf(z)
z, p
```

```python
print ('Porcentaje de estudiantes que obtuvieron igual o menor de 70:' , round(p* 100 , 2 ))
```

- ¿Qué porcentaje de estudiantes obtuvieron menos de 80?:

```python
z = (80-mu)/sigma
p = norm.cdf(z)
z, p
```

```python
print ('Porcentaje de estudiantes que obtuvieron igual o menor de 80:' , round(p* 100 , 2 ))
```

- ¿Qué porcentaje de estudiantes obtuvieron mas de 80?:

```python
z = (80-mu)/sigma
p = norm.cdf(z)
display(z, p),
p=1-p
display(p)
```

```python
print ('Porcentaje de estudiantes que obtuvieron mas de 80:' , round(p* 100 , 2 ))
```

- - ¿Cual es la calificaciòn que saco del 98% acumulado?

```python
z = norm.ppf(0.98)
z = (z*sigma)+mu
z
```

```python
print( "La calificaciòn para la cual 98% estudiantes son más bajos:" , round(z, 2))
```
